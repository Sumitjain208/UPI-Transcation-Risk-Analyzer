```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("transactions.csv")
```

```python
df.sample(12)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Timestamp']=pd.to_datetime(df['Timestamp'])
```

```python
df.info()
```

```python
df.sample(4)
```

```python
df['date_month']=df['Timestamp'].dt.month
```

```python
df['date_day']=df['Timestamp'].dt.day
```

```python
df['date_hour']=df['Timestamp'].dt.hour
df.head(50)
```

```python
df['is_night']=df['date_hour'].apply(lambda x:1 if x>=23 or x<=5 else 0)
```

```python
df.sample(3)
```

```python
df.drop(columns=['Transaction ID',
                 'Sender Name',
                 'Receiver Name'],inplace=True)
                 
```

```python
df['Sender_bank']=df['Sender UPI ID'].apply(lambda x:x.split('@')[1])
```

```python
df['Receiver_bank']=df['Receiver UPI ID'].apply(lambda x:x.split('@')[1])
```

```python
df.drop(columns=['Sender UPI ID','Receiver UPI ID'],inplace=True)
```

```python
df.head()
```

```python
df['Sender_bank'].unique()
```

```python
df['Receiver_bank'].unique()
```

```python
!pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Status']=le.fit_transform(df['Status'])
df['Sender_bank']=le.fit_transform(df['Sender_bank'])
df['Receiver_bank']=le.fit_transform(df['Receiver_bank'])
```

```python
df.head()
```

```python
df['is_fraud']=(
    (df['Amount (INR)']>7000) & (df['Status']==0) & (df['is_night']==1)).astype(int)
```

```python
print(df['is_fraud'].value_counts())
```

```python
!pip install seaborn
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
sns.boxplot(df['Amount (INR)'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

df['Amount (INR)']=scaler.fit_transform(df[['Amount (INR)']])
```

```python
sns.distplot(df['Amount (INR)'])
```

```python
df.drop('Timestamp',axis=1,inplace=True)
```

```python
!pip install imbalanced-learn
```

```python
X=df.drop('is_fraud',axis=1)
```

```python
y=df['is_fraud']
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
X_resembled,y_resembled=smote.fit_resample(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_resembled,y_resembled,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
model.fit(X_train,y_train)
    
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print('Accuracy:',accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.show()
```

```python
importance=model.feature_importances_
features_names=X.columns
```

```python
plt.figure(figsize=(10,6))
plt.bar(features_names,importance)
plt.xticks(rotation=45)
plt.show()
```

```python
import pickle
pickle.dump(model,open('fraud_model.pkl','wb'))
```
